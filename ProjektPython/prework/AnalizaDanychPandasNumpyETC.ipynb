```python
import pandas as pd
df = pd.DataFrame.from_records([( "A", 1,31),( "A", 2,41),( "A", 1,44), ( "A", 3,51), ("B",1,515), ("B",2,51),( "C", 2,155),( "C", 134)], columns = ["user", "prod_in_basket","basket_value"])


#groupby tworzy grupowanie po kolumnach, można podać listę kolumn jeżeli chcemy grupować po kilku kolumnach
df_groups = df.groupby(by = "user")
#w obrębie grup możemy policzyć różne statystyki po kolumnach które nie służycły do grupowania (w naszym przypadku "user")
#mamy dostępne funkcje typu sum, min, max... funkcja sum (itp) zwracają series, stąd trzeba je przekonwertować z powrotem do ramki danych (to_frame)
#kolumna z agregatami ma domyślnie nazwę kolumny na podstawie której ją policzono, można zmienić nazwę funkcją .rename
basket_size = df_groups['prod_in_basket'].sum().to_frame().rename(columns = {'prod_in_basket':'prod_in_basket_sum'})
print("*"*20,"\nbasket_size\n", basket_size)
basket_value = df_groups['basket_value'].sum().to_frame().rename(columns = {'basket_value':'basket_value_sum'})
print("*"*20,"\nbasket_value\n", basket_value)
basket_value_max = df_groups['basket_value'].max().to_frame().rename(columns = {'basket_value':'basket_value_max'})
print("*"*20,"\nbasket_value_max\n", basket_value_max)

#agregaty można połączyć w jedną ramkę funkcją cocnat, należy pamiętać o axis = 1, w przeciwnym razie każdy z agregatów zostanie potraktowany jako oddzielny wiersz
print(pd.concat([basket_size,basket_value,basket_value_max], axis = 1))
#w przykładzie wyżej kolumna grupująca traktowana jest jak indeks (nie jest to zwykła kolumna), żeby przywrócić user do zwykłych kolumn należy wykonać reset_index
print(pd.concat([basket_size,basket_value,basket_value_max], axis = 1).reset_index())
```
